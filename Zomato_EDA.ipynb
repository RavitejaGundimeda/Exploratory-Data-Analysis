```python
final_df.columns
```

```python
ratings = final_df.groupby([ 'Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

##Observations
1.  When rating > 4.5 and rating <= 4.9 ratings are Excellent
2.  When rating > 4.0 and rating < 4.4 ratings are Very Good
3.  When rating > 3.5 and rating < 3.9 ratings are Good
4.  When rating > 3.0 and rating < 3.4 ratings are Average
5.  When rating > 2.5 and rating < 3.0 ratings are Avergae
6.  When rating > 2.0 and rating < 2.4 ratings are Poor


```python
ratings.head()
```

```python
import matplotlib as plt
plt.rcParams['figure.figsize'] = (12,6)
```

```python
sns.barplot(x="Aggregate rating", y = "Rating Count", data = ratings )
```

```python
sns.barplot(x="Aggregate rating", y = "Rating Count", data = ratings,hue = 'Rating color',palette=['gray','red','orange','yellow','green','green'])
```

#Observations:
1. Not rated count is high
2. Max ratings are between 2.5 and 3.4


```python
sns.countplot(x='Rating color', data = ratings, palette = ['gray','red','orange','yellow','green','green']) #this shows the frequency of the colors not the count
```
