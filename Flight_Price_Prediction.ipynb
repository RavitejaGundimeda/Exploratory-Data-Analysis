#Goal is to handle null values, date format, and label encoding.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
dfTrain = pd.read_excel('/content/Data_Train.xlsx') #load train dataset
dfTest = pd.read_excel('/content/Test_set.xlsx') #load test dataset
```

```python
df = dfTrain.append(dfTest) #append the test data to train
```

```python
df.tail()
```

```python
df.info()
```

###Creating derived features from Date of Journey.

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(3)
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis = 1, inplace=True)
```

```python
df.head(1)
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

###Check for null values

```python
df.isnull().sum()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
#df.head(3)
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Minutes'] = df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
#df.head(3)
```

```python
df.drop('Dep_Time',axis =1, inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'].unique()
```

##Null values are replaced with 1 based on the domain knowledge.

```python
# Assuming df is your DataFrame
df['Total_Stops'] = pd.to_numeric(df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4}), errors='coerce').fillna(1) #Replace non numeric values with NaN and fills NaN with 1
```

```python
df['Total_Stops'].unique()
```

```python
df.head()
```

```python
df.drop('Route',axis = 1, inplace=True)
```

```python
#df.info()
```

##Converting hours to minutes.

```python
df['Duration'] = df['Duration'].apply(lambda x: int(x.split('h')[0]) * 60 + int(x.split(' ')[-1][0]) if 'h' in x else int(x.split('m')[0]))
```

```python
labelEncoder = LabelEncoder()
df['Airline'] = labelEncoder.fit_transform(df['Airline'])
df['Source'] = labelEncoder.fit_transform(df['Source'])
df['Destination'] = labelEncoder.fit_transform(df['Destination'])
```
